```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Distribution 1: Geometric Distribution

geometric = np.random.geometric(0.75,100)

#Vertical line at the mean
plt.axvline(geometric.mean(), color='green', linestyle='solid', linewidth=2)

#Vertical line one SD above mean
plt.axvline(geometric.mean()+geometric.std(), color='purple', linestyle='dashed', linewidth=2)

#Vertical line one SD below mean
plt.axvline(geometric.mean()-geometric.std(), color='purple', linestyle='dashed', linewidth=2)


plt.hist(geometric)
plt.show()

#Since this histogram is asymmetrical, or skewed, the mean and standard deviation are not the best measures of 
#central tendency and variability.  
```

```python
#Distribution 2 - Hypergeometric Distribution

hypergeometric = np.random.hypergeometric(10,10,10,100)

#Vertical line at the mean
plt.axvline(hypergeometric.mean(), color='green', linestyle='solid', linewidth=2)

#Vertical line one SD above mean
plt.axvline(hypergeometric.mean()+hypergeometric.std(), color='purple', linestyle='dashed', linewidth=2)

#Vertical line one SD below mean
plt.axvline(hypergeometric.mean()-hypergeometric.std(), color='purple', linestyle='dashed', linewidth=2)

plt.hist(hypergeometric)
plt.show()

#Since this histogram is skewed left, the mean and standard deviation are not the best measures of 
#central tendency and variability. 
```

```python
#Distribution 3 - Poisson Distribution

poisson = np.random.poisson(10,100)

#Vertical line at the mean
plt.axvline(poisson.mean(), color='green', linestyle='solid', linewidth=2)

#Vertical line one SD above mean
plt.axvline(poisson.mean()+poisson.std(), color='purple', linestyle='dashed', linewidth=2)

#Vertical line one SD below mean
plt.axvline(poisson.mean()-poisson.std(), color='purple', linestyle='dashed', linewidth=2)

plt.hist(poisson)
plt.show()

#Since this histogram is skewed, the mean and standard deviation are not the best measures of 
#central tendency and variability.
```

```python
#Distribution 4 - Negative Binomial Distribution

negative_binomial = np.random.negative_binomial(3,0.4,100)

#Vertical line at the mean
plt.axvline(negative_binomial.mean(), color='green', linestyle='solid', linewidth=2)

#Vertical line one SD above mean
plt.axvline(negative_binomial.mean()+negative_binomial.std(), color='purple', linestyle='dashed', linewidth=2)

#Vertical line one SD below mean
plt.axvline(negative_binomial.mean()-negative_binomial.std(), color='purple', linestyle='dashed', linewidth=2)

plt.hist(negative_binomial)
plt.show()

#Since this histogram is skewed right, the mean and standard deviation are not the best measures of 
#central tendency and variability.
```

```python
#Additional 

norm1 = np.random.normal(5, 0.5, 100)
norm2 = np.random.normal(10, 1, 100)

plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.hist(norm1)
plt.title('Mean=5, std=0.5')

#Vertical lines at mean, one SD above mean, and one SD below mean
plt.axvline(norm1.mean(), color='green', linestyle='solid', linewidth=2)
plt.axvline(norm1.mean()+norm1.std(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(norm1.mean()-norm1.std(), color='purple', linestyle='dashed', linewidth=2)

plt.subplot(1,2,2)
plt.hist(norm2)
plt.title('Mean=10, std=1')

#Vertical lines at mean, one SD above mean, and one SD below mean
plt.axvline(norm2.mean(), color='green', linestyle='solid', linewidth=2)
plt.axvline(norm2.mean()+norm2.std(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(norm2.mean()-norm2.std(), color='purple', linestyle='dashed', linewidth=2)

plt.show()
```

```python
new_norm = norm1 + norm2

#Vertical lines at mean, one SD above mean, and one SD below mean
plt.axvline(new_norm.mean(), color='green', linestyle='solid', linewidth=2)
plt.axvline(new_norm.mean()+new_norm.std(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(new_norm.mean()-new_norm.std(), color='purple', linestyle='dashed', linewidth=2)

plt.hist(new_norm)
plt.show()

#This histogram is slightly more subtle, but there is some skew.  Therefore, the descriptive statistics aren't the best measures.  
```
