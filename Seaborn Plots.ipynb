```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mass_shootings = pd.read_csv('mass_shootings.csv')
mass_shootings.head()
```

```python
# Choose one variable and plot that variable four different ways
# Variable: US Fatalities from Mass Shootings from 1966 to 2017

# 1. Visualizing Fatalities with Histogram
plt.figure(figsize=(8, 8))

sns.distplot(mass_shootings.Fatalities, kde=False)
plt.title('Histogram: US Fatalities from Mass Shootings, 1966-2017')
plt.show()
```

Pros: Easy to compare data, easily understand the range, overview of distribution and frequency, identify outliers, can identify mode and mean <br>
Cons: Not very precise(can't accurately determine the amount of input data)

```python
# 2. Visualizing Fatalities with box plot
plt.figure(figsize=(8, 8))
sns.boxplot(y=mass_shootings.Fatalities)
plt.title('Box Plot: US Fatalities from Mass Shootings, 1966-2017')
plt.show()
```

Pros: Clear summary of the spread: median, interquantile range, range(max/min values), identify outliers, shows skew <br>
Cons: Doesn't identify exact values/precise details or amount of input data, mean and mode not identifiable

```python
# 3. Visualizing Fatalities with violinplot
sns.violinplot(x=mass_shootings.Fatalities)
plt.title('Violinplot: US Fatalities from Mass Shootings, 1966-2017')
plt.show()
```

Pros: Provides overview of summary statistics, shows distribution of data and skew, shows mode/bimodality <br>
Cons: No exact way to determine accurate input data

```python
# 4. Visualizing Fatalities with Density Plot
sns.kdeplot(mass_shootings['Fatalities'])
plt.title('Density Plot: US Fatalities from Mass Shootings, 1966-2017')
plt.show()
```

Pros: Shows general distribution, skew, mode, outliers <br>
Cons: No exact values or amout of input data, harder to digest than a histogram that has bins/can't see distribution of bins

```python
# Choose two continuous variables, and plot them three different ways
# Variables: Fatalities and Total victims

# 1. Visualizing Fatalities and Total victims with scatterplot
sns.set(style="white")

g = sns.lmplot(x='Total victims', y='Fatalities', data=mass_shootings, fit_reg=False, scatter_kws={'alpha':0.4})
plt.title('Scatterplot: US Fatalities by Total Victims in Mass Shootings')
plt.show()
```

Pros: Every data point is represented, can see if correlation exists, clustering, shows relationship between data <br>
Cons: Hard to determine distribution, scale/plotted points overlap, outliers affect scale, no easily discernible summary stats

```python
# 2. Visualizing Fatalities and Total victims with Regplot

sns.set(style="white")

g = sns.regplot(x='Total victims', y='Fatalities', data=mass_shootings, scatter_kws={'alpha':0.4})
plt.title('Regression plot: US Fatalities by Total Victims in Mass Shootings')
plt.show()
```

Pros: Every data point is represented, can see if correlation exists and has regression line, clustering, shows relationship between data <br>
Cons: Hard to determine distribution, scale/plotted points overlap, outliers affect scale, no easily discernible summary stats

```python
# 3. Visualizing Fatalities and Total victims with Histograms

sns.distplot(mass_shootings['Fatalities'], color='black', label='Fatalities', kde=False)
sns.distplot(mass_shootings['Total victims'], color='blue', label='Total victims', kde=False)
plt.title('Histograms: Distribution US Fatalities and Total Victims in Mass Shootings')
plt.legend()
plt.show()
```

Pros: Easily understand the range, overview of distribution and frequency, identify outliers, can identify mode and mean <br>
Cons: Not the best way to compare data that overlaps, not very precise(can't accurately determine the amount of input data)

```python
# Choose one continuous variable and one categorical variable, and plot them six different ways
# Variables: US Fatalities and Race of shooter

# 1. Visualizing Race of shooter(categorical) and Fatalities (continuous) with bar plot
plt.figure(figsize=(10, 10))
sns.set(style='darkgrid')

sns.barplot( x='Fatalities', y='Race', data=mass_shootings)
plt.title('Barplot: Number of US Fatalies by Race of Shooter')
plt.show()
```

Pros: Good way to make comparisons between groups, can be easily manipulated(ex. must start at 0), too many bars make plot overwhelming and less useful <br>
Cons: Can't determine distribution, can't determine the number of datapoints , don't show relationships

```python
# 2. Visualizing Race of shooter(categorical) and Fatalities (continuous) with Point Plot
sns.set(style='darkgrid')

sns.factorplot( x='Fatalities', y='Race', data=mass_shootings, size=10, ci=95, join=False)
plt.title('Pointplot: US Fatalities from Mass Shootings by Race of Shooter, 1966-2017')
plt.show()
```

Pros: Use less space to convey same info as barplots <br>
Cons: Can't determine distribution of categorical data, can't determine the number of datapoints, don't show relationships

```python
# 3. Visualizing Race of shooter(categorical) and Fatalities (continuous) with Boxplot
plt.figure(figsize=(10, 10))
sns.boxplot(x=mass_shootings.Fatalities, y=mass_shootings.Race)
plt.title('Boxplot: US Fatalities from Mass Shootings by Race of Shooter, 1966-2017')
plt.show()
```

Pros: Clear summary of the spread: median, interquantile range, range(max/min values), identify outliers, shows skew, shows comparisons between categorical data <br>
Cons: Doesn't identify exact values/precise details or amount of input data, mean and mode not identifiable, too much categorical data makes it hard to make apt comparisons

```python
# 4. Visualizing Race of shooter(categorical) and Fatalities (continuous) with Violinplot
f, ax = plt.subplots(figsize=(7, 23))
sns.violinplot(x=mass_shootings.Fatalities, y=mass_shootings.Race)
plt.title('Violinplot: US Fatalities from Mass Shootings by Race of Shooter, 1966-2017')
plt.show()
```

Pros: Provides overview of summary statistics, shows distribution of data and skew, shows mode/bimodality, makes comparisons of summary stats and distribution between categorical data <br>
Cons: No exact way to determine accurate number of input datapoints, too much categorical data can skew scale and make comparisons difficult depending on range and distribution

```python
# 5. Visualizing Race of shooter(categorical) and Fatalities (continuous) with Strip plot
plt.figure(figsize=(8, 8))
sns.stripplot(x=mass_shootings.Fatalities, y=mass_shootings.Race, jitter=True)
plt.title('Strip Plot: US Fatalities from Mass Shootings by Race of Shooter, 1966-2017')
plt.show()
```

Pros: Can see outliers, good way to compare a few groups, illustrates and compares ranges aptly <br>
Cons: Hard to compare with more than a few categorical groups, doesn't show distribution or summary stats, difficult to read

```python
# 6. Visualizing Race of shooter(categorical) and Fatalities (continuous) with Swarm plot
plt.figure(figsize=(8, 8))
sns.swarmplot(x=mass_shootings.Fatalities, y=mass_shootings.Race)
plt.title('Swarm Plot: US Fatalities from Mass Shootings by Race of Shooter, 1966-2017')
plt.show()
```

Pros: Can see outliers, good way to compare a few groups, illustrates and compares ranges aptly, better way to see distribution than strip plot <br>
Cons: Hard to compare with many categorical groups, hard to read, for categorical data that isn't represented by many points it's hard to make any meaningful inferences

```python
#Find a dataset with at least four continuous variables and one categorical variable. 
#Create one master plot that gives insight into the variables and their interrelationships, including:
#Probability distributions, Bivariate relationships, Whether the distributions or the relationships vary across groups
#Accompany your plot with a written description of what you see.
mass_shootings_edit = mass_shootings[['Fatalities', 'Total victims', 'Injured', 'Policeman Killed', 'Gender']]
sns.pairplot(mass_shootings_edit)
plt.show()
```

Probability Distributions: Fatalies and Policeman Killed are both skewed right, whereas Injured and Total Victims each have one bar <br>
Bivariate Relationships: The relationships vary across groups.  All continuous variables against Policemen Killed have a weak relationship, if any at all.  Fatalities, Total victims, and Injured datapoints are tightly clustered on the left hand axes with the exception of a couple outliers. As the Total victims variable increases, the Injured and Fatalities variables trend in the same way - generally as the total victims increase, the injured and fatalities increase as well.
